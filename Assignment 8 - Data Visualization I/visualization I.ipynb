```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
# Kernel Density Estimate plot (a smooth line) should not be included in the plot if false
#  a histogram is a graphical representation of the distribution of a dataset

sns.histplot(df['Fare'],kde=True)    
```

```python
sns.histplot(df['Fare'],kde=False)
```

```python
sns.histplot(df['Fare'],kde=True, bins = 30)
# Each bin represents a specific range of values.
```

```python
# A hexbin plot divides the space into hexagonal bins, and the color in each bin represents
# the number of points that fall into that bin. It's a useful way to visualize the density of points in a scatter plot.
sns.jointplot(x='Age', y='Fare', data=df, kind='hex')
```

```python
# The resulting plot will have scatter plots in the off-diagonal positions and histograms
# or kernel density plots along the diagonal. Each point in the scatter plots represents a data point, 
# and the patterns can help identify trends or correlations between variables.

sns.pairplot(df)
```

```python
df = df.dropna()
```

```python
sns.pairplot(df, hue='Sex')
```

```python
# This type of plot is useful for quickly understanding the distribution of categorical variables. 
# In your case, you'll see how many occurrences there are for each gender in the 'Sex' column.

sns.countplot(x='Sex', data=df)
```
