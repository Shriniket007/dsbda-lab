```python
df["math score"].replace(np.NaN, df["math score"].astype("float").mean(), inplace=True)
df.isnull().sum()
```

```python
df["reading score"].replace(np.NaN, df["reading score"].astype("float").mean(), inplace=True)
df["writing score"].replace(np.NaN, df["writing score"].astype("float").mean(), inplace=True)
df.isnull().sum()
```

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr

outliers = ((df < lower_bound) | (df > upper_bound)).any(axis = 1)
df_new = df[~outliers]
df_new.head()
```
